```python
# перемножения значений двух игральных кубиков
n = 0
m = 0
for i in range(1,7):
    for o in range(1,7):
        m += 1
        if i * o <= 10:
            n += 1
print(n, m)
```

```python
# формула бернулли
# sp.comb(n, k)
# n - общее кол-во вариантов
# k - сколько должно выпасть 
```

```python
# 1
c = sp.comb(5, 3)
x = (0.4**3 * 0.6**2 * c).round(4)
x
```

```python
# 2
p = 0.6
q = 1 - p
n = 5
d = 0
for i in range(5):
    x = (p ** i * q ** (n-i)) * sp.comb(n, i)
    d += x
    #print(x)
print(d.round(3))
#print(1-d)
```
