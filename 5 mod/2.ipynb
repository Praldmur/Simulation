```python
from scipy import special as sp
```

```python
p = 0.1
q = 1 - p
n = 30
```

```python
d = 0
for i in range(2):
    x = (p ** i * q ** (n-i)) * sp.comb(n, i)
    d += x 
    print(x)
print(d)
print(1-d)
```
