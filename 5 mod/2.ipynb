Вопрос:
Какова вероятность того, что из этих трех пользователей ровно двое кликнут на уведомление?

Системный шаблон: Схема БернуллиКогда у тебя есть $n$ одинаковых испытаний, в каждом из которых успех происходит с вероятностью $p$, а неуспех с вероятностью $q = 1-p$, формула для нахождения ровно $k$ успехов выглядит так:$$P_n(k) = C_n^k \cdot p^k \cdot q^{n-k}$$Разберем на твоем примере (3 пуша, 2 клика):$C_n^k$ — это «количество путей».Ты уже умеешь это считать! Это количество способов выбрать, какие именно 2 пользователя из 3 кликнут.Мы это считали через «рукопожатия» или факториалы: $C_3^2 = 3$.(Пути: К-К-И, К-И-К, И-К-К).$p^k$ — это «вероятность успехов».У нас 2 клика, каждый по 0.1. Значит: $0.1 \cdot 0.1 = 0.1^2 = 0.01$.$q^{n-k}$ — это «вероятность неудач».Оставшийся 1 пользователь не кликнул (вероятность 0.9). Значит: $0.9^1 = 0.9$.Собираем конструктор:$$P = 3 \cdot 0.01 \cdot 0.9 = 0.027$$

```python
c = sp.comb(3, 2)
x = (0.1**2 * 0.9**1 * c).round(4)
x
```
