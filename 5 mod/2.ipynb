Системный шаблон: Схема БернуллиКогда у тебя есть $n$ одинаковых испытаний, в каждом из которых успех происходит с вероятностью $p$, а неуспех с вероятностью $q = 1-p$, формула для нахождения ровно $k$ успехов выглядит так:$$P_n(k) = C_n^k \cdot p^k \cdot q^{n-k}$$Разберем на твоем примере (3 пуша, 2 клика):$C_n^k$ — это «количество путей».Ты уже умеешь это считать! Это количество способов выбрать, какие именно 2 пользователя из 3 кликнут.Мы это считали через «рукопожатия» или факториалы: $C_3^2 = 3$.(Пути: К-К-И, К-И-К, И-К-К).$p^k$ — это «вероятность успехов».У нас 2 клика, каждый по 0.1. Значит: $0.1 \cdot 0.1 = 0.1^2 = 0.01$.$q^{n-k}$ — это «вероятность неудач».Оставшийся 1 пользователь не кликнул (вероятность 0.9). Значит: $0.9^1 = 0.9$.Собираем конструктор:$$P = 3 \cdot 0.01 \cdot 0.9 = 0.027$$

```python
def bern(n, k, p):
    """
    Вычисляет вероятность по формуле Бернулли.
    n — количество испытаний
    k — количество успехов
    p_success — вероятность успеха в одном испытании
    """
    result = sp.comb(n, k) * (p ** k) * ((1 - p) ** (n - k))
    return result
```

Оценка риска фичи «Чаевые курьеру»Ты анализируешь новую фичу — «Чаевые курьеру в один клик» (когда сервис автоматически предлагает добавить фиксированную сумму к заказу, а компания берет с этого микро-комиссию за эквайринг).  
Ты раскатил фичу на часть пользователей и выяснил, что чистый доход компании с одного заказа распределяется следующим образом:
- Сценарий «Щедрость» (Вероятность 20%): Пользователь оставляет крупные чаевые, доход компании — 500 рублей.

Твое задание:Напиши Python-скрипт, который рассчитает риски этой фичи. Тебе нужно посчитать:
- Математическое ожидание дохода ($M[X]$).
- Дисперсию ($D[X]$) с учетом веса (вероятностей) каждого сценария.
- Стандартное отклонение ($\sigma$).   
Важный нюанс: Так как у каждого исхода своя вероятность, обычный np.var() от списка [500, 100, 0] тут не подойдет (он подумает, что шансы равны). Используй «ручной» алгоритм через списки или цикл, реализуя формулу:$$D[X] = \sum (x_i - M[X])^2 \cdot p_i$$Выведи результат красивой f-строкой. Посмотрим, насколько высок риск у этой фичи!

```python
x1 = 500
p1 = 0.2
x2 = 100
p2 = 0.5
x3 = 0
p3 = 0.3

m_x = x1*p1 + x2*p2 + x3*p3
d_x = (x1-m_x)**2*p1 + (x2-m_x)**2*p2 + (x3-m_x)**2*p3
sd = round((d_x**0.5),2)
print(f'мат.ожидания равны {int(m_x)}')
print(f'дисперсия равна {int(d_x)}')
print(f'стандартное отклонение равно {int(sd)}')
```

```python
X = np.array([500, 100, 0])
P = np.array([0.2, 0.5, 0.3])

# Проверка на сумму вероятностей
if np.isclose(np.sum(P), 1.0):
    # Умножение происходит поэлементно, а np.sum() всё складывает
    m_x = np.sum(X * P)
    
    # NumPy сам вычтет m_x из каждого элемента X и возведет в квадрат
    d_x = np.sum(((X - m_x) ** 2) * P)
    
    sd = np.sqrt(d_x)

    print("--- Расчет через NumPy ---")
    print(f"Мат. ожидание: {int(m_x)}")
    print(f"Дисперсия: {int(d_x)}")
    print(f"Стандартное отклонение: {sd:.2f}")
```
