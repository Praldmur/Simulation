Задание 10  
Брошены две игральные кости. Есть следующие события:  
- А = {выпадение шестерки на первой кости}
- В = {сумма выпавших очков равна 7}
- С = {выпадение тройки на второй кости}   
Какие пары событий являются независимыми? 
- В и С
- А и С

Довольно очевидно, что A и C друг от друга не зависят - ведь бросок одной кости никак не меняет результат на другой. Но как быть с A и B, а также B и C? Проверим это через формулу независимости событий:  
P(A∩B)=P(A)P(B)  
- P(A)=1/6 (так как всего граней шесть, а нас интересует всего одно значение)  
- P(B)=6/36=1/6 (всего комбинаций двух кубиков 36, а условию соответствуют варианты 6 и 1, 5 и 2, 4 и 3, 3 и 4, 2 и 5, 1 и 6)
- P(A∩B)=1/36 (если выпала шестерка, то событие B произойдёт только для одной комбинации из 36 - 6 и 1)  
Итого: 1/36=1/6∗1/6, что соответствует независимости событий. То же самое справедливо для события C.  
Значит, все эти события независимы.

Пример задачи  
Ситуация:
Ты отправил 3 пуш-уведомления разным пользователям. Ты знаешь по прошлому опыту, что вероятность того, что пользователь кликнет на пуш (CTR), составляет 0.1 (10%) для каждого. Предположим, что пользователи действуют независимо друг от друга.

Вопрос:
Какова вероятность того, что из этих трех пользователей ровно двое кликнут на уведомление?

Системный шаблон: Схема Бернулли: Когда у тебя есть $n$ одинаковых испытаний, в каждом из которых успех происходит с вероятностью $p$, а неуспех с вероятностью $q = 1-p$, формула для нахождения ровно $k$ успехов выглядит так:$$P_n(k) = C_n^k \cdot p^k \cdot q^{n-k}$$Разберем на твоем примере (3 пуша, 2 клика):$C_n^k$ — это «количество путей».Ты уже умеешь это считать! Это количество способов выбрать, какие именно 2 пользователя из 3 кликнут.Мы это считали через «рукопожатия» или факториалы: $C_3^2 = 3$.(Пути: К-К-И, К-И-К, И-К-К).$p^k$ — это «вероятность успехов».У нас 2 клика, каждый по 0.1. Значит: $0.1 \cdot 0.1 = 0.1^2 = 0.01$.$q^{n-k}$ — это «вероятность неудач».Оставшийся 1 пользователь не кликнул (вероятность 0.9). Значит: $0.9^1 = 0.9$.Собираем конструктор:$$P = 3 \cdot 0.01 \cdot 0.9 = 0.027$$

```python
# формула бернулли
# sp.comb(n, k)
# n - общее кол-во вариантов
# k - сколько должно выпасть 
```

```python
c = sp.comb(3, 2)
x = (c * 0.1**2 * 0.9**1).round(4)
x
```

```python
def bern(n, k, p):
    """
    Вычисляет вероятность по формуле Бернулли.
    n — количество испытаний
    k — количество успехов
    p_success — вероятность успеха в одном испытании
    """
    result = sp.comb(n, k) * (p ** k) * ((1 - p) ** (n - k))
    return result
```

Оценка запуска новой подписки (Три исхода)
Команда хочет запустить новый премиум-транспорт в сервисе доставки (например, доставку дронами или супер-экспресс). Разработка и поддержка этой инфраструктуры в пересчете на одного привлеченного пользователя обойдется компании в 800 рублей (это наши фиксированные затраты, то есть минус).

Ты изучил предзаказы и поведение пользователей на тесте и выделил три сценария:

Сценарий «Успех» (Вероятность 10%): Пользователь влюбляется в фичу, оформляет VIP-подписку и приносит компании 5000 рублей дохода.

Сценарий «Средне» (Вероятность 50%): Пользователь пользуется фичей изредка, покупает разовые доставки и приносит 1000 рублей дохода.

Сценарий «Провал» (Вероятность 40%): Пользователь заходит, ничего не понимает, уходит и приносит 0 рублей.

Вопрос: Выгодно ли компании запускать эту фичу в производство? Каково чистое математическое ожидание прибыли с одного пользователя?

```python
#Матожидание
c = 800
x1 = 5000
p1 = 0.1
x2 = 1000
p2 = 0.5
x3 = 0
p3 = 0.4

# Проверяем корректность вероятностей
if round(p1 + p2 + p3, 4) == 1.0:
    m_x = x1*p1 + x2*p2 + x3*p3
    net_m_x = m_x - c
    
    # Используем f-строки для красивого вывода
    if net_m_x > 0:
        print(f"мат.ожидания ({int(net_m_x)}) > 0, выкатываем")
    else:
        print(f"мат.ожидания ({int(net_m_x)}) < 0, не выкатываем")
else: 
    print('Ошибка: сумма вероятностей не равна 1')
```

Вариант Б: «Реклама + Премиум» (Гибридная монетизация)

Фича более простая, затраты в пересчете на пользователя: 20 рублей.

Твоё задание:
Напиши Python-скрипт (используя f-строки, проверку суммы вероятностей round(p1 + ..., 4) == 1.0 и условия if/else), который:

Рассчитает чистое матожидание прибыли для Варианта А.

Рассчитает чистое матожидание прибыли для Варианта Б.

Сравнит их и автоматически выведет на экран финальную рекомендацию для продакта (какой вариант выгоднее и на сколько).

```python
# сценарий А
c = 30
x1 = 600
p1 = 0.2
x2 = 0
p2 = 0.8

if round(p1 + p2, 4) == 1.0:
    m_a = x1*p1 + x2*p2
    net_m_a = m_a - c
else: 
    print('Ошибка: сумма вероятностей в вар.А не равна 1')
    
# сценарий Б
c = 20
x1 = 2000
p1 = 0.05
x2 = 50
p2 = 0.4
x3 = 0
p3 = 0.55

if round(p1 + p2 + p3, 4) == 1.0:
    m_b = x1*p1 + x2*p2 + x3*p3
    net_m_b = m_b - c
else: 
    print('Ошибка: сумма вероятностей в вар.Б не равна 1')

if net_m_a > net_m_b:
    print(f"мат.ожидания ({int(net_m_a)}) > ({int(net_m_b)}), выкатываем вар.А")
elif net_m_a < net_m_b:
    print(f"мат.ожидания ({int(net_m_b)}) > ({int(net_m_a)}), выкатываем вар.Б")
else:
    print(f"мат.ожидания равны ({int(net_m_a)}) = ({int(net_m_b)})")
```

Оценка риска фичи «Чаевые курьеру»Ты анализируешь новую фичу — «Чаевые курьеру в один клик» (когда сервис автоматически предлагает добавить фиксированную сумму к заказу, а компания берет с этого микро-комиссию за эквайринг).  
Ты раскатил фичу на часть пользователей и выяснил, что чистый доход компании с одного заказа распределяется следующим образом:
- Сценарий «Щедрость» (Вероятность 20%): Пользователь оставляет крупные чаевые, доход компании — 500 рублей.
- Сценарий «Стандарт» (Вероятность 50%): Пользователь оставляет обычные чаевые, доход компании — 100 рублей.  
- Сценарий «Экономия» (Вероятность 30%): Пользователь кликает «Не оставлять чаевые», доход компании — 0 рублей.   
Твое задание:Напиши Python-скрипт, который рассчитает риски этой фичи. Тебе нужно посчитать:
- Математическое ожидание дохода ($M[X]$).
- Дисперсию ($D[X]$) с учетом веса (вероятностей) каждого сценария.
- Стандартное отклонение ($\sigma$).   
Важный нюанс: Так как у каждого исхода своя вероятность, обычный np.var() от списка [500, 100, 0] тут не подойдет (он подумает, что шансы равны). Используй «ручной» алгоритм через списки или цикл, реализуя формулу:$$D[X] = \sum (x_i - M[X])^2 \cdot p_i$$Выведи результат красивой f-строкой. Посмотрим, насколько высок риск у этой фичи!

```python
x1 = 500
p1 = 0.2
x2 = 100
p2 = 0.5
x3 = 0
p3 = 0.3

m_x = x1*p1 + x2*p2 + x3*p3
d_x = (x1-m_x)**2*p1 + (x2-m_x)**2*p2 + (x3-m_x)**2*p3
sd = round((d_x**0.5),2)
print(f'мат.ожидания равны {int(m_x)}')
print(f'дисперсия равна {int(d_x)}')
print(f'стандартное отклонение равно {int(sd)}')
```

```python
X = np.array([500, 100, 0])
P = np.array([0.2, 0.5, 0.3])

# Проверка на сумму вероятностей
if np.isclose(np.sum(P), 1.0):
    # Умножение происходит поэлементно, а np.sum() всё складывает
    m_x = np.sum(X * P)
    
    # NumPy сам вычтет m_x из каждого элемента X и возведет в квадрат
    d_x = np.sum(((X - m_x) ** 2) * P)
    
    sd = np.sqrt(d_x)

    print("--- Расчет через NumPy ---")
    print(f"Мат. ожидание: {int(m_x)}")
    print(f"Дисперсия: {int(d_x)}")
    print(f"Стандартное отклонение: {sd:.2f}")
```

Представь, что мы в Самакате внедрили новый интерфейс корзины (версия Б). Мы хотим увеличить средний чек.  
- Мы знаем, что в старой версии (А) средний чек равен 1000 рублей, а сигма ($\sigma$) равна 200 рублей.

Продакт кричит: «Ура! Чек вырос на 50 рублей! Катим на всех!».Но ты, как аналитик, смотришь на размер выборки и понимаешь, что нужно посчитать, а не случайность ли это.  
Давай посчитаем Z-score для этого изменения, но уже для среднего значения (в статистике для этого сигму делят на корень из размера выборки, но давай пока представим, что мы смотрим на одного конкретного типичного пользователя новой группы).  
Если $M[X] = 1000$, $\sigma = 200$, а новое значение $X = 1050$:
- Чему равен Z-score для этого пользователя?
- Входит ли этот результат в рамки «обычного шума» (правило 2-3 сигм) или это уже железный повод утверждать, что новый дизайн кардинально изменил поведение человека?

```python
z = (1050 - 1000) / 200
z
```

- Посчитай новый Z-score по формуле выше (сначала найди корень из $n$, потом раздели сигму на этот корень, и уже на получившееся число дели разницу в 50 рублей).
- Сравни получившийся Z-score с критическим значением $1.96$.
- Каков будет вердикт для продакта теперь? Изменилось ли что-то от того, что пользователей стало больше?
